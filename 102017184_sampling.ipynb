```python
import numpy as np
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/AnjulaMehto/Sampling_Assignment/main/Creditcard_data.csv'
data = pd.read_csv(url)
data.head()
```

```python
class_freq = data['Class'].value_counts()
print(f"Number of classes: {len(class_freq)}")
print("Class frequencies:")
print(class_freq)
```

```python
#Converting the dataset into balanced dataset

from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler

# Separate features and target
x = data.drop('Class', axis=1)
y = data['Class']

# Instantiate oversampler and undersampler
oversampler = RandomOverSampler()
undersampler = RandomUnderSampler()

# Resample using both oversampling and undersampling
x_over, y_over = oversampler.fit_resample(x, y)
x_resampled, y_resampled = undersampler.fit_resample(x_over, y_over)

# Print the number of samples in each class
print("Number of samples in each class after resampling:")
print(y_resampled.value_counts())
```

```python
x_resampled
```

```python
new_data = pd.concat([x_resampled, y_resampled], axis=1)
new_data
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

Sampling the data in the next few cells using different sampling techniques.

```python
#1

# Simple random sampling
n1 = int((1.96*1.96 * 0.5*0.5)/(0.05**2))
random_sample = new_data.sample(n=n1, random_state=42)
print(random_sample)
```

```python
#2

# Systematic sampling
interval = 2
n2 = (1.96**2 * 0.5 * (1-0.5)) / 0.05**2
systematic_sample = new_data.iloc[::interval]
systematic_sample
```

```python
#3

# Stratified sampling
n3 = int((1.96**2 * 0.5*0.5)/((0.05)**2))
strata = new_data.groupby('Class')

# Sample 2 rows from each stratum
stratified_sample = strata.apply(lambda x: x.sample(n3))
stratified_sample
```

```python
#4

# Cluster sampling
import numpy as np

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=10, random_state=42).fit(new_data)
cluster_assignments = kmeans.labels_

n4 = int((1.96**2 * 0.5*0.5)/((0.05/10)**2))

# Select the clusters you want to include in the sample
selected_clusters = [0, 2, 4, 5, 8]

cluster_series = pd.Series(cluster_assignments)

# Create the new DataFrame containing only the rows from the selected clusters
cluster_sample = new_data[cluster_series.isin(selected_clusters)]

# Print the resulting DataFrame
(cluster_sample)
```

```python
#5

# Convenience sampling
n5 = int((1.96**2 * 0.5*0.5)/((0.05)**2))
convenience_sample = pd.concat([new_data.head(380), new_data.tail(380)])
convenience_sample
```

Applying different Machine Learning Models to the above created samples.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import ExtraTreesClassifier

model1 = LogisticRegression(random_state=42,max_iter=500)
model2 = DecisionTreeClassifier(random_state=42)
model3 = RandomForestClassifier(random_state=42)
model4 = SVC(random_state=42)
model5 = ExtraTreesClassifier(random_state=42)

# Define a dictionary to hold the sampling techniques and models
samplers = {
    'Sampling1': random_sample,
    'Sampling2': systematic_sample,
    'Sampling3': stratified_sample,
    'Sampling4': cluster_sample,
    'Sampling5': convenience_sample,
}
models = {
    'Model1': model1,
    'Model2': model2,
    'Model3': model3,
    'Model4': model4,
    'Model5': model5,
}
```

```python
# Evaluate each model on each sampling technique

from sklearn.metrics import accuracy_score

results = {}
for sampler_name, sampler in samplers.items():
    if sampler_name == 'Sampling1':
        n = n1
    elif sampler_name == 'Sampling2':
        n = n2
    elif sampler_name == 'Sampling3':
        n = n3
    elif sampler_name == 'Sampling4':
        n = n4
    else:
        n = n5
    
    # Undersample or oversample the training data
    x_resampled, y_resampled = undersampler.fit_resample(x_train, y_train)
    
    # Limit the resampled data to the sample size
    if len(x_resampled) > n:
        x_resampled = x_resampled[:n]
        y_resampled = y_resampled[:n]
    
    for model_name, model in models.items():
        # Train the model on the resampled data
        model.fit(x_resampled, y_resampled)
        
        # Make predictions on the test data
        y_pred = model.predict(x_test)
        
        # Calculate the accuracy score
        accuracy = accuracy_score(y_test, y_pred)
        
        # Add the accuracy score to the results dictionary
        if model_name in results:
            results[model_name][sampler_name] = accuracy
        else:
            results[model_name] = {sampler_name: accuracy}
```

```python
# Print the results
print('Results:')
print('        Sampling1   Sampling2   Sampling3   Sampling4   Sampling5')
for model_name, model_results in results.items():
    print(model_name, end='')
    for sampler_name in samplers.keys():
        if sampler_name in model_results:
            print(f'    {model_results[sampler_name]:.4f}   ', end='')
        else:
            print('              ', end='')
    print()
```
